```python
from google.colab import drive
import zipfile, os

# === 1. Mount your Google Drive ===
drive.mount('/content/drive')
current_folder = '/content/drive/MyDrive/lesson'  # Your folder

# === 2. Install gdown (if not installed) ===
!pip install gdown

# === 3. Set the shared file ID and destination filename ===
file_id = '1Ie21sSgEMBM9YVzSTU2W6gWqRCXL73yx'  # Your shared file ID
destination = os.path.join(current_folder, 'dataset-12.zip')

# === 4. Download the file using gdown (handles virus warning) ===
import gdown

# Construct the Google Drive URL
url = f"https://drive.google.com/uc?id={file_id}"

# Download the file
gdown.download(url, destination, quiet=False)

# === 5. Unzip the downloaded file ===
with zipfile.ZipFile(destination, 'r') as zip_ref:
    zip_ref.extractall(current_folder)

print(f"✅ File downloaded and unzipped in: {current_folder}")
```
